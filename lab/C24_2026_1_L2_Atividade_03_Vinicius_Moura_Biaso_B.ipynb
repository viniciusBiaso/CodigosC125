### <font color='cyan'>Q4: O conjunto de dados possui características numéricas? Se sim, escreva seu(s) nome(s) e classifique-a(s) como contínua(s) e/ou discreta(s).</font>
- #### <font color='cyan'>Observação: considere que uma caraterística numérica é discreta se possui menos que 50 valores únicos.</font>

```python
#classificação se possui caracteristicas numericas
colunas_numericas = df.select_dtypes(include=['number']).columns.tolist()
print(f"Colunas numéricas encontradas: {colunas_numericas}")

classificacao = []

for coluna in colunas_numericas:
    #float continua, int discreto
    if df[coluna].dtype == 'float64' or df[coluna].dtype == 'float32':
        tipo = "Contínua"
    else:
        tipo = "Discreta"

    classificacao.append({"Coluna": coluna, "Tipo": tipo})

df_tipos = pd.DataFrame(classificacao)
print(df_tipos)
```

## <font color='green'>**2.5 Identificação de Problemas nos Dados**</font>

### <font color='cyan'> Q16: Inspecione o conjuntos de dados para verificar se há:</font>
- #### <font color='cyan'> Dados Ausentes;</font>
- #### <font color='cyan'> Valores Atípicos (Outliers);</font>
- #### <font color='cyan'> Valores Duplicados;</font>
### <font color='cyan'> Reporte sua(s) descoberta(s) a seguir.</font>

```python
# Verifica o total de valores ausentes por coluna
dados_ausentes = df.isnull().sum()
print("Valores ausentes por coluna:")
print(dados_ausentes)

#verifica valores duplicados
valores_duplicados = df.duplicated().sum()
print(f"valores duplicados: {valores_duplicados}")

Q1 = df['flow_duration'].quantile(0.25)
Q3 = df['flow_duration'].quantile(0.75)
IQR = Q3 - Q1

limite_inferior = Q1 - 1.5 * IQR
limite_superior = Q3 + 1.5 * IQR

outliers = df[(df['flow_duration'] < limite_inferior) | (df['flow_duration'] > limite_superior)]
print(f"Quantidade de outliers encontrados: {len(outliers)}")
```

### <font color='cyan'> Q17: Caso problemas tenham sido encontrados, proponha soluções para eles.</font>
### <font color='cyan'> Exiba o conjunto de dados antes e depois de aplicar as soluções, enfatizando as mudanças.</font>

```python
#retirar valores duplicados
df_sem_duplicados = df.drop_duplicates()

#correcao de outliers
mediana = df['flow_duration'].median()

# Onde for menor que o limite inferior OU maior que o superior, substitui pela mediana
df.loc[(df['flow_duration'] < limite_inferior) | (df['flow_duration'] > limite_superior), 'flow_duration'] = mediana
```
